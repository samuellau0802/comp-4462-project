```python
# Read JSON data from files
import json


with open('src/data/Aggregated_Data_v2.json', 'r') as f:
    old = json.load(f)

with open('src/data/Combined_Data_2008_2023.json', 'r') as f:
    new = json.load(f)
```

```python
combined_data = {}

# Function to merge two dictionaries and check for overlapping keys
def merge_dicts(dict1, dict2, path=None):
    if path is None:
        path = []
    for key in dict2:
        if key in dict1:
            if isinstance(dict1[key], dict) and isinstance(dict2[key], dict):
                merge_dicts(dict1[key], dict2[key], path + [key])
            else:
                print(f"Overlapping key: {'.'.join(path + [key])}")
                dict1[key] = dict2[key]
        else:
            dict1[key] = dict2[key]

# Function to find unique countries in each dictionary
def find_unique_countries(dict1, dict2):
    countries1 = set(dict1.keys())
    countries2 = set(dict2.keys())
    unique_in_dict1 = countries1 - countries2
    unique_in_dict2 = countries2 - countries1
    if unique_in_dict1:
        print(f"Countries only in the first JSON: {', '.join(unique_in_dict1)}")
    if unique_in_dict2:
        print(f"Countries only in the second JSON: {', '.join(unique_in_dict2)}")
```

```python
new = {key: val for key, val in new.items() if key in set(correct_countries)}
```

```python
# Function to merge two dictionaries and check for overlapping keys
def merge_dicts(dict1, dict2, path=None):
    if path is None:
        path = []
    for key in dict2:
        mapped_key = map_country_name(key)
        if mapped_key in dict1:
            if isinstance(dict1[mapped_key], dict) and isinstance(dict2[key], dict):
                merge_dicts(dict1[mapped_key], dict2[key], path + [mapped_key])
            else:
                if dict1[mapped_key] != dict2[key]:
                    print(f"Overlapping key: {'.'.join(path + [mapped_key])}")
                    print(f"Value in dict1: {dict1[mapped_key]}")
                    print(f"Value in dict2: {dict2[key]}")
                dict1[mapped_key] = dict2[key]
        else:
            dict1[mapped_key] = dict2[key]

# Combine the JSON data
combined_data = {}
merge_dicts(combined_data, old)
merge_dicts(combined_data, new)
```

```python
# Save combined_data to a JSON file
with open('src/data/countries_data_2008_2023.json', 'w') as json_file:
    json.dump(combined_data, json_file, indent=4)
```
